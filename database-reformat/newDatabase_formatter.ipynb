```python
# replaces underscores with spaces:
df.columns = df.columns.str.replace('_', ' ')
df.head(6)
```

```python
#truncates the zip code

if "ZIP" in df:
        df["ZIP"] = df["ZIP"].str[:5]    
if "zip" in df:
        df["zip"] = df["zip"].str[:5]
df.head(6)
```

```python
#capitalize the country and city for searching

if "Country" in df:
        df['Country'] = df['Country'].str.title()
if "country" in df:
        df['country'] = df['country'].str.title()
if "COUNTRY" in df:
        df['COUNTRY'] = df['COUNTRY'].str.title()
if "City" in df:
        df['City'] = df['City'].str.title()
df.head(6)
```

```python
#searches the package pycountry for the alpha-3 code for the country
#uses both common names and official names so it can detect more effectively
if "Country" in df:
    def country_code_search(country):
        try:
            result = pycountry.countries.search_fuzzy(country)
        except Exception:
            return np.nan
        else:
            return result[0].alpha_3
    alpha_3 = {country: country_code_search(country) for country in df["Country"].unique()}
    df["Country"] = df["Country"].map(alpha_3)
df.head(6)
```
