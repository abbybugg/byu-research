```python
# Single month smaller period of time search trends data compiler. This allows you to access google search trends 
# data and then: export it to a database, export to csv, or export as json with vertices and edges 
##########################################################
import csv
import pandas as pd
import json
from json import JSONEncoder
from pandas import DataFrame
from pytrends.request import TrendReq
import numpy as np
# import time
# import ast

# Login to Google. Only need to run this once, the rest of requests will use the same session.
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
# List of search terms:

keyword = (['coronavirus'])
location = ('US-UT')
timeframe = ('2020-05-01 2020-05-30')

filename = ('monthdata.txt')
```

```python
# Extract data from Google Trends:

# Create payload and capture API tokens.
pytrend.build_payload(keyword, cat=0, timeframe=timeframe,geo=location,gprop='')

# Create dataframe of interest over time
df = pytrend.interest_over_time()

#Format dataframe:
df = df.reset_index() 
df = df.reset_index()
df = df.drop(columns = ['date', 'isPartial'])
df.columns=['Date','Search Popularity']
df['Location'] = location
```

```python
#EXPORTING TO DATABASE:
##########################################################
import sqlite3
from pandas import DataFrame

conn = sqlite3.connect('DATABASE.db')
c = conn.cursor()
conn.commit()

df.to_sql("Data",conn,if_exists='append', index = False)
conn.close()
```

```python
# EXPORTING TO CSV FILE
##########################################################
df.to_csv('month_search_data')
```

```python
# EXPORT AS JSON FILE WITH VERTICES/EDGES
##########################################################
# Create vertices: 
df_list = df.values.tolist()
vertices = { i : df_list[i] for i in range(0, len(df_list) ) }

# And edges:
edges = []
for i in range(1,len(df)):
    edge = [str(i-1),str(i)]
    edge = edges.append(edge)

finaldict = {"Data": { "Vertices": vertices, "Edges": edges}}
with open(filename, 'a+') as outfile:
    json.dump(finaldict, outfile)
```
