```python
import csv
import pandas as pd
import json
from json import JSONEncoder
from pandas import DataFrame
from pytrends.request import TrendReq
import numpy as np
import sqlite3
```

```python
# SQL Query for the specific data we want:

sql = 'SELECT Date,"Search: coronavirus", "New Cases", "New Deaths" FROM Search_Data'
filename = ('finaldata.txt')
```

```python
# Create dataframe of table in database:
conn = sqlite3.connect('DATABASE.db')
c = conn.cursor()
conn.commit()

df = pd.read_sql(sql,con=conn)
```

```python
# Define vertices and edges with whole year in one dictionary:
df_list = df.values.tolist()
vertices = { i : df_list[i] for i in range(0, len(df_list)) }
edges = []
for i in range(0,len(df)-1):
    edge = [str(i),str(i+1)]
    edge = edges.append(edge)
    finaldict = {"Data": { "Vertices": vertices, "Edges": edges}}
    with open(filename, "a") as outfile:
         json.dump(finaldict, outfile)
            
```
